```python
#Data Cleaning  
# Return a new Data Frame with no empty cells 

import pandas as pd

new = pd.read_csv('data.csv')

new_data = new.dropna()

print(new_data.to_string())
```

```python
# Remove all rows with NULL values: 
new = pd.read_csv('data.csv')

new.dropna(inplace = True)

print(new.to_string())
```

```python
# Replace empty values 

# The fillna() method allows us to replace empty cells with a value:

new = pd.read_csv('data.csv')

new.fillna(130, inplace = True)

print(new)
```

```python
# Replace only for specified columns: 

new = pd.read_csv('data.csv')

new["Calories"].fillna(130, inplace = True)

print(new)
```

```python
# Replace using mean, median or mode
# mean() median() and mode() methods

new = pd.read_csv('data.csv')

x = new["Calories"].mean()

new["Calories"].fillna(x, inplace = True)

print(new)
```

```python
new = pd.read_csv('data.csv')

x = new["Calories"].median()

new["Calories"].fillna(x, inplace = True)

print(new)
```

```python
# Calculate the MODE, and replace any empty values with it: 

new = pd.read_csv('data.csv')

x = new["Calories"].mode()[0]

new["Calories"].fillna(x, inplace = True)

print(new)
```

```python
# to_datetime() method 

new = pd.read_csv('data.csv')

new['Calories'] = pd.to_datetime(new['Calories'])

print(new.to_string())
```
